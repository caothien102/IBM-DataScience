```python
import yfinance as yf
import matplotlib.pyplot as plt

# Define the make_graph function
def make_graph(stock_data, title):
    """
    Function to plot stock data.
    Arguments:
    - stock_data: A DataFrame containing stock data with 'Date' and 'Close' columns
    - title: The title for the graph
    """
    plt.figure(figsize=(10,6))
    plt.plot(stock_data['Date'], stock_data['Close'], label="Close Price", color='blue')
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()  # Ensures everything fits without overlap
    plt.show()
# Call make_graph to plot the data with a title
make_graph(tesla_stock_data, "Tesla Stock Price Over Time")
make_graph(gamestop_stock_data, "Gamestop Stock Price Over Time")

```
